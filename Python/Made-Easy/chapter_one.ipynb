    1.1. Введение в Data Science.
Многие годы важным во всем мире считали 2 веши: духовное и маериальное.

А вот информацию или данные никто в расчет не брал.

Данные произвели революцию в технологиях в неменьшей степени, чем электричество. Сегодня происходит цифровая революция, которая чем-то похожа на самоисполняющееся пророчество. Она словно двигатель, который питает сам себя.

На данный момент цифровым миром управляют данные, хотя часто не показывая это.
Подобное утверждение может показаться слишком надуманным или натянутым, но нет — это наша реальность, здесь и сейчас.

Когда Ozon и Yandex market на своих торговых площадках показывают вам товары, которые могут вас заинтересовать, именно данные лежат в основе анализации ваших предподчений.

Когда политики во всем мире задействуют социальные сети и используют собранные данные для социального профилирования, данные лежат в основе принятия важных решений, например о том, какой кандидат, где и как сможет выиграть выборы.

Когда Yandex Maps сообщает вам, насколько сильные пробки на дороге, именно данные программа собирает чтобы это определить.

Когда аналитики рынка акций прогнозируют будущие котировки на фондовом рынке, они используют анализ данных.

На данный момент сфера анализа данных очень востребованно поскольку данные и аналитика данных окружает нас повсюду, мы уже начали производить больше данных за секунду чем прошые поколения за всю свою жизнь.

Проще говоря, Data Science — это наука об изучении данных для того, чтобы данные обрели смысл и дали нам полезную информацию. Эта наука не касается какойто конкретной предметной области, а являет собой сочетание различных дисциплин, связанных с анализом данных и выдаче «умных» результатов и решений, основанных на этих данных.

Data Science — это междисциплинарная область, которая помогает нам извлекать знания из структурированных или неструктурированных данных (рис. 1.1).

        1.1.1. Зачем нам Data Science?

Аналитика данных была создана нам в помошь, что бы мы точнее принимали решения, состовляли прогнозы. Эта наука помогает нам:

1. Задавать правильные вопросы;

2. Находить основную причину проблемы;

3. Находить закономерности среди, на первый взгляд, хаотичных необработанных
данных;

4. Создавать модели для предиктивного анализа;

5. Визуализировать и отображать результаты с помощью графиков, информационных панелей;

6. Проектировать AI;

7. Определять лояльность клиентов с помощью анализа настроений;

8. Принимать более качественные и быстрые решения;

9. Рекомендовать правильный продукт нужному клиенту для развития нашего бизнеса.

        1.1.2. История Data Science

Data Science на самом деле существует довольно давно но не теряет своей актуальности, про Data Science можно сказать как старый Джинн в новой бутылке.

Сбор данных и составление на их основе значимых выводов — далеко не новое занятие. Ведь если подумать и не углублятся в детали, то Data Science был ещё с доисторических времён. Например, цифры, используемые в древности пастухами, которым нужно было пересчитать свой скот, — это и были первые наборы данных.

            1.1.2.1. Данные в древности

    Около 18 000 г. до н. э.

Самый ранний способ хранения и анализа данных человеком был 18 000 г. до н. э. Кость Ишанго, обнаруженная в 1960 г. на территории современной Уганды, считается одним из самых ранних доисторических свидетельств хранения данных. Палеолитические племена делали зарубки на палках или костях, чтобы вести счет припасов и торговли. Они сравнивали палки и насечки, выполняя тем самым элементарные расчеты, что позволяло предугадывать, например, на сколько хватит их запасов еды.

    Около 2400 г. до н. э.

Счеты — первое устройство для работай с данными тоесть с числами были сделаны в Вавилоне. Интерестно но  в тоже время появились первые библиотеки с книжками - для хранения информации. 

            1.1.2.2. Появление статистики

    1663 год

Джон Граунт - учёный из Лондона, который первый провел официыльный экспиримент по статическому анализу данных в 1663 году. Он предложил что с помощью имеющихся данных о смертности, он сможет вывести алгоритм который на раней стадии будет предупреждать о чуме, сгубившей 40% населения Европы.

    1865 год

В Энциклопедии «Энциклопедии коммерческих и деловых анекдотов» Ридарда Миллар Девенс впервые описывается такой термин как бизнес-аналитика, описывая им то как банкир Генри Фернезе добился преимущества перед конкурентами, собирая, анализируя и структурируя информацию с его бизнеса. Это в первый раз показало миру как использовать аналитику в комерчиском бизнесе.

    1880 год

Во время переписи населения в США у Бюро возникла проблема из-за огромного количество данных из-за чего было предсказано что на обработку данных потребуется 8 лет, а в 1890 г. 10 лет. Эти данные даже не успеют обработать до следующей переписи 1900 г. Но в 1881 году в Бюро нанимают молодого инденера - Германа Холлерита который создал нечто, что поже назавут "Табулирующей машиной Холлерита". С помощью перфокарт он сократил 10 лет обработки данных до трех месяцев и занял свое место в истории как отец-основатель современных автоматизированных вычислений. Основанная им компания впоследствии станет называться IBM

    1926 год

Никола Тесла во время интервью сказал что при правильном применении беспроводных технологий «всю планету можно превратить в "огромный мозг", что можно сказать и случилась, Снтевая паутина расползлась по всей земле. И все на ней является её частичкой, например наши мобильные телефоны.

            1.1.2.4. Появление больших центров обработки данных.

    1965 год

            1.1.2.5. Появление Интернета.

    1991 год

Тим Бернерс-Ли - Ученый IT предсказал появление интернета.

    1997 год

Появился первый поиковой сервис "Google Search", и в течении 20 лет название "Google" обрела нарицательный характер.

В том же году запускается журнал «Data Mining and Knowledge Discovery», причем такой порядок этих двух терминов в его названии отражает то, что термин Data Mining (сбор данных) стал более популярен, чем Knowledge Discovery (извлечение информации из больших баз данных). 

    2000 год

16 октября 2000 г. был создан на c++ язык Python 2.0 с множеством новых функций, включая сборщик мусора с обнаружением циклических ссылок и поддержку Unicode

    2011 год

В отчете McKinsey говорится, что к 2018 г. США крайне нужны будут специалисты в области обработки данных, а такие проблемы, как конфиденциальность, безопасность и интеллектуальная собственность, должны быть решены до того, как большие данные раскроют весь свой потенциал.

    2012 год

В сентябре 2012 г. Том Дэвенпорт и Д. Дж. Патил публикуют статью «Data Scientist: The Sexiest Job of the 21st Century» («Специалист по обработке данных: самая сексуальная работа XXI века») в журнале Harvard Business Review.

    2014 год

Наблюдается спрос на мобильные устройства, мобильный способ общения стали вытиснять пк. 88 % руководителей предприятий, опрошенных GE совместно с Accenture, считают, что аналитика больших данных является основным приоритетом для их бизнеса.
